# **Perceptron : Introduction**

> **What is it ?**

> A perceptron is the fundamental building block of artificial neural networks. It's a simple model of a biological neuron that takes several inputs, processes them, and produces a single output.

image.png

Features:
1. Inputs(x1,x2,x3...,xn):
These are the input features fed into the perceptron. They represent the data points you want the perceptron to learn from.

2. Weights(W1,W2,W3,...Wn):
Each input (xᵢ) is multiplied by a corresponding weight (Wᵢ). These weights determine the importance of each input in the calculation of the weighted sum.

Generally, the larger the absolute value of a weight, the more influence the corresponding input has on the perceptron's output.

3.  Weighted Sum (Σ):
This is where each input is multiplied by its corresponding weight, and then all these products are summed together.

(w0*x + w1*y + w2*z) = z

4. Activation Function (S-shaped symbol)
The weighted sum (z) is then passed through an activation function. The activation function introduces non-linearity to the model, allowing it to learn complex patterns.
It also squashes the output of the weighted sum into a specific range, say (-1,1) or (0,1)

> Various Activation Functions:
  * Sigmoid Function
  * Tanh Function
  * Step Function
  * ReLU

5. Output(Y):
The output of the activation function (often denoted as 'y' or 'ŷ') is the final output of the perceptron.

#### Adding bias(b) to weighted sum
 Mathetmatically, if we have inputs x1,x2,..xn and their corresponding weights w1,w2,,..wn. Then weighted sum,

z = ( x1* w1 + x2 * w2 ..+ xn*wn)+b

 where 'b' is bias,

 Think of it as an extra constant term that allows the activation function to be shifted.It allows the perceptron to activate even when all the input features are zero.

```python
# Code example : Perceptron Algorithm!
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
```

```python
df.head()
```

```python
sns.scatterplot(x = df['cgpa'],y = df['resume_score'],hue = df['placed'])
```

```python
X = df.iloc[:,:-1]
y = df['placed']
```

```python
from sklearn.linear_model import Perceptron

p = Perceptron()
```

```python
p.fit(X,y)
```

```python
# Value of weights w1 and w2
p.coef_
```

```python
# Value of bias
p.intercept_
```

```python
# mlxtend (machine learning extensions) is a lib in py.
# plot_decision_regions ;used for visualizing the decision boundaries of classification models in 2D or 3D feature spaces.
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf =p, legend = 2)
```
